### Importing Necessary Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

### Loading the Dataset

```python
# Load the dataset
data = pd.read_csv('total_data_na.csv')
df=pd.DataFrame(data)
```

### Checking for Missing Values

```python
df.isnull() # used to check null values
null_Val = df.isnull().sum() # returning total null value of each column
print("Number of Null Values: ", null_Val)
```

### Data Cleaning: Converting Columns to Numeric
Object columns are converted to numeric types where applicable, replacing any errors with NaN.


```python
# Step 1: Data Cleaning
# Convert object columns to numeric, replacing errors with NaN
df['Avg.x'] = pd.to_numeric(df['Avg.x'], errors='coerce')
df['Avg.y'] = pd.to_numeric(df['Avg.y'], errors='coerce')
df['SR.y'] = pd.to_numeric(df['SR.y'], errors='coerce')
df['Econ'] = pd.to_numeric(df['Econ'], errors='coerce')
```

### Handling Non-Numeric Values
Non-numeric values in specific columns are coerced into numeric types to ensure compatibility with analysis.


```python
# Handle any non-numeric values (like '-') in the selected features
for col in ['Mat.x', 'Inns.x', 'HS', 'Avg.x', 'SR.x','Mat.y', 'Inns.y', 'Ov', 'Avg.y', 'Econ', 'SR.y', 'Wkts']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Imputing Missing Values
Missing values in critical columns are filled using median or mean imputation methods to handle missing data effectively.


```python
# Impute missing values: Avg.x with median, Avg.y and SR.y with mean
df['Avg.x'] = df['Avg.x'].fillna(df['Avg.x'].median())
df['Avg.y'] = df['Avg.y'].fillna(df['Avg.y'].mean())
df['SR.y'] = df['SR.y'].fillna(df['SR.y'].mean())
df['Econ'] = df['Econ'].fillna(df['Econ'].mean())
df['Ov'] = df['Ov'].fillna(df['Ov'].mean())
```

### Feature Selection
Relevant features are selected for predicting runs scored and wickets taken, ensuring that all features are numeric.


```python
# Ensure all selected features are numeric
features_runs = df[['Mat.x', 'Inns.x', 'HS', 'Avg.x', 'SR.x']].fillna(0)  # or use imputation methods
labels_runs = df['Runs.x']  # Using 'Runs.x' as the target label

# Select features for wicket prediction
features_wickets = df[['Mat.y', 'Inns.y', 'Ov', 'Avg.y', 'Econ', 'SR.y']]
labels_wickets = df['Wkts']  # Using 'Wkts' as the target label
```

### Splitting the Data
The data is split into training and testing sets for building machine learning models to predict runs and wickets.


```python
# Split the data into training and testing sets
X_runs_train, X_runs_test, y_runs_train, y_runs_test = train_test_split(features_runs, labels_runs, test_size=0.2, random_state=42)
X_wickets_train, X_wickets_test, y_wickets_train, y_wickets_test = train_test_split(features_wickets, labels_wickets, test_size=0.2, random_state=42)
```

### Training the Linear Regression Model
Linear regression models are trained on the training data for predicting both runs and wickets.


```python
# Initialize and train the linear regression model
model_runs = LinearRegression()
model_runs.fit(X_runs_train, y_runs_train)
model_wickets = LinearRegression()
model_wickets.fit(X_wickets_train, y_wickets_train)
```

### Predicting Values
Predictions are made on the test data using the trained linear regression models.


```python
#predicting the values
y_runs_pred = model_runs.predict(X_runs_test)
y_wickets_pred = model_wickets.predict(X_wickets_test)
```

### Evaluate the model using Mean Squared Error (MSE) and R-squared (R²).

```python
#Evaluate the model
mse_runs = mean_squared_error(y_runs_test, y_runs_pred)
mse_wickets = mean_squared_error(y_wickets_test, y_wickets_pred)

r2_runs = r2_score(y_runs_test, y_runs_pred)
r2_wickets = r2_score(y_wickets_test, y_wickets_pred)

print("Mean Squared Error (Runs):", mse_runs)
print("Mean Squared Error (Wickets):", mse_wickets)

print("R-squared (Runs):", r2_runs)
print("R-squared (Wickets):", r2_wickets)
```

### Collect user input for batting statistics and predict runs using the `get_user_input_runs()` function with error handling for valid data entry.

```python
# Taking user input for Batting 

def get_user_input_runs():
    try:
        Mat_x = int(input("Enter number of matches played (Mat.x): "))
        Inns_x = int(input("Enter number of innings played (Inns.x): "))
        HS = int(input("Enter highest score (HS): "))
        Avg_x = float(input("Enter batting average (Avg.x): "))
        SR_x = float(input("Enter strike rate (SR.x): "))
        return pd.DataFrame({
            'Mat.x': [Mat_x],
            'Inns.x': [Inns_x],
            'HS': [HS],
            'Avg.x': [Avg_x],
            'SR.x': [SR_x]
        })
    except ValueError:
        print("Invalid Input! Please enter the values again.")
        return get_user_input_runs()

new_data_runs = get_user_input_runs()
new_prediction_runs = model_runs.predict(new_data_runs)
print("Predicted runs based on input values: ", new_prediction_runs[0])
```

### Collect user input for wickets statistics and predict runs using the `get_user_input_wickets()` function with error handling for valid data entry.

```python
# Taking user input for bowling
# Function to get user input for wickets prediction
def get_user_input_wickets():
    try:
        Mat_y = int(input("Enter number of matches played (Mat.y): "))
        Inns_y = int(input("Enter number of innings bowled (Inns.y): "))
        Ov = float(input("Enter number of overs bowled (Ov): "))
        Avg_y = float(input("Enter bowling average (Avg.y): "))
        Econ = float(input("Enter economy rate (Econ): "))
        SR_y = float(input("Enter bowling strike rate (SR.y): "))
        return pd.DataFrame({
            'Mat.y': [Mat_y], 'Inns.y': [Inns_y], 'Ov': [Ov],
            'Avg.y': [Avg_y], 'Econ': [Econ], 'SR.y': [SR_y]
        })
    except ValueError:
        print("Invalid Input! Please enter the values again.")
        return get_user_input_wickets()
    
new_data_wickets = get_user_input_wickets()
new_prediction_wickets = model_wickets.predict(new_data_wickets)
print("Predicted wickets based on input values: ", new_prediction_wickets[0])
```
